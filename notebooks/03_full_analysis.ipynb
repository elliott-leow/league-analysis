# causal graph structures across ranks

generates and displays the learned bayesian network cpdag for each rank tier.


```python
import sys
from pathlib import Path
from IPython.display import Image, display
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# add project root to path
sys.path.insert(0, str(Path.cwd().parent))

from src import ges, visualize
from src.config import ALL_RANKS, FIGURES_DIR

print("✓ loaded")
print(f"ranks: {ALL_RANKS}")
```

---

## platinum rank

combat-first strategy: kill advantages drive objective control


```python
rank = 'Platinum'

# load model and generate graph
try:
    result = ges.load_ges_result(rank)
    print(f"loaded {rank} model: {result['n_edges']} edges")
    
    # generate visualization
    output_path = FIGURES_DIR / f'cpdag_{rank}.png'
    visualize.plot_cpdag(
        result['edges'],
        result['variables'],
        title=f'{rank} Rank - Causal Structure',
        output_file=output_path,
        layout='spring'
    )
    plt.close()  # close matplotlib figure to prevent double display
    print(f"✓ generated {output_path.name}")
    
    # display saved image
    display(Image(filename=str(output_path), width=800))
    
except Exception as e:
    print(f"error: {e}")
```

## all ranks comparison

2x2 grid showing all learned causal structures


```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# create 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(16, 16))
fig.suptitle('Causal Graph Structures Across All Ranks', fontsize=16, fontweight='bold')

ranks = ['Platinum', 'Diamond', 'Master', 'Elite']
positions = [(0, 0), (0, 1), (1, 0), (1, 1)]

for rank, (row, col) in zip(ranks, positions):
    img_path = FIGURES_DIR / f'cpdag_{rank}.png'
    if img_path.exists():
        img = mpimg.imread(str(img_path))
        axes[row, col].imshow(img)
        axes[row, col].axis('off')
        axes[row, col].set_title(f'{rank} Rank', fontsize=14, fontweight='bold')
    else:
        axes[row, col].text(0.5, 0.5, f'{rank}\nImage not found', 
                           ha='center', va='center', fontsize=12)
        axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```
